```julia

# Using a for loop and int16

a = Matrix{Int16}(undef, 100, 100)
b = Matrix{Int16}(undef, 100, 100)
c = Matrix{Int16}(undef, 100, 100)

for x in 1:1000000
    rand!(a, 0:999)
    rand!(b, 0:999)
    c .= a .+ b
end
```

```julia
# Using a for Loop and int32
# Found it to be slower than int16

a = Matrix{Int32}(undef, 100, 100)
b = Matrix{Int32}(undef, 100, 100)
c = Matrix{Int32}(undef, 100, 100)

for x in 1:1000000
    rand!(a, 0:999)
    rand!(b, 0:999)
    c .= a .+ b
end
```

```julia
# Defining a function to take advantage of JIT
# I didn't notice a difference despite this being a supposed advantage of using Julia over Python

using Random


a = Matrix{Int16}(undef, 100, 100)
b = Matrix{Int16}(undef, 100, 100)
c = Matrix{Int16}(undef, 100, 100)

function loop_add(n)
    for x in 1:n
        rand!(a, 0:999)
        rand!(b, 0:999)
        c .= a .+ b
    end
end

loop_add(1000000)
```

```julia
# Defining a function to take advantage of JIT
# Attempting to use all of my threads
# It didn't make any difference, I am not sure if using mutiple threads is advantageous or possible within coLab


using Random, Base.Threads

a = Matrix{Int16}(undef, 100, 100)
b = Matrix{Int16}(undef, 100, 100)
c = Matrix{Int16}(undef, 100, 100)

function loop_add_parallel(n)
    @threads for x in 1:n
        rand!(a, 0:999)
        rand!(b, 0:999)
        c .= a .+ b
    end
end

loop_add_parallel(1000000)
```
