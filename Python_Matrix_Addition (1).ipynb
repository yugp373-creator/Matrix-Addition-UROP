```python
# used dtype to test the effect of int16 vs int32 to see if int16 would be faster


import numpy as np

rng = np.random.default_rng()

for x in range(1, 1000001):
  a = np.random.randint(0, 1000, size=(100, 100), dtype=np.int32)
  b = np.random.randint(0, 1000, size=(100, 100), dtype=np.int32)
  c = a + b
```

```python
# found int16 to be faster, likely because it uses less memory and the numbers within the matrix addition program are small

import numpy as np

rng = np.random.default_rng()

for x in range(1, 1000001):
    a = rng.integers(0, 1000, size=(100, 100), dtype=np.int16)
    b = rng.integers(0, 1000, size=(100, 100), dtype=np.int16)
    c = a + b

```
